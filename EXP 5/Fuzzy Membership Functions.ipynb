# Importing Necessary Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
```

# Define universe of discourse

```python
x = np.arange(0, 11, 1)
```

# Membership Functions

# 1. Triangular Membership Function

```python
A = fuzz.trimf(x, [0, 0, 5])       # Low
B = fuzz.trimf(x, [3, 5, 7])       # Medium
C = fuzz.trimf(x, [5, 10, 10])     # High
```

```python
plt.figure(figsize=(16, 12))

# Plot Triangular Membership Functions
plt.subplot(3, 2, 1)
plt.plot(x, A, label='Low (A)', color='blue')
plt.fill_between(x, 0, A, color='blue', alpha=0.2)
plt.title('Triangular Membership Function: Low')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()

plt.subplot(3, 2, 2)
plt.plot(x, B, label='Medium (B)', color='green')
plt.fill_between(x, 0, B, color='green', alpha=0.2)
plt.title('Triangular Membership Function: Medium')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()

plt.subplot(3, 2, 3)
plt.plot(x, C, label='High (C)', color='red')
plt.fill_between(x, 0, C, color='red', alpha=0.2)
plt.title('Triangular Membership Function: High')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()
```

# 2. Trapezoidal Membership Function

```python
D = fuzz.trapmf(x, [0, 2, 8, 10])  # Mid-Range
E = fuzz.trapmf(x, [0, 3, 7, 10])  # Wide Range
```

```python
# Plot Trapezoidal Membership Functions
plt.subplot(3, 2, 4)
plt.plot(x, D, label='Mid-Range (D)', color='orange', linestyle='--')
plt.fill_between(x, 0, D, color='orange', alpha=0.2)
plt.title('Trapezoidal : Mid-Range')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()
```

# 3. Gaussian Membership Function

```python
F = fuzz.gaussmf(x, mean=5, sigma=1.5)  # Gaussian centered at 5 with spread 1.
```

```python
# Plot Gaussian Membership Function
plt.subplot(3, 2, 5)
plt.plot(x, F, label='Gaussian (F)', color='purple')
plt.fill_between(x, 0, F, color='purple', alpha=0.2)
plt.title('Gaussian Membership Function')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()
```

# 4. Bell Membership Function 

```python
G = fuzz.gbellmf(x, a=2, b=4, c=5) 
```

```python
# Plot Generalized Bell Membership Function
plt.subplot(3, 2, 6)
plt.plot(x, G, label='Bell (G)', color='brown')
plt.fill_between(x, 0, G, color='brown', alpha=0.2)
plt.title('Generalized Bell Membership Function')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

# 5. Sigmoid Membership Function

```python
H = fuzz.sigmf(x, 5, 1)  # Sigmoid MF centered at 5 with steepness 1
```

```python
# Visualize the Sigmoid Membership Function
plt.figure(figsize=(10, 6))
plt.plot(x, H, label='Sigmoid MF (H)', color='purple')
plt.fill_between(x, 0, H, color='purple', alpha=0.2)
plt.title('Sigmoid Membership Function')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()
plt.show()
```
