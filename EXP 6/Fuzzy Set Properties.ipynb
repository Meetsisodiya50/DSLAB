# Importing Necessary Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

# Universe of Discourse for Fuzzy Operation

```python
x = np.arange(0, 11, 1)

# Define fuzzy variables with triangular membership functions
A = fuzz.trimf(x, [0, 0, 5])  # Fuzzy set A: Low
B = fuzz.trimf(x, [3, 5, 7])  # Fuzzy set B: Medium
C = fuzz.trimf(x, [5, 10, 10]) # Fuzzy set C: High

# Plot membership functions
plt.figure(figsize=(12, 8))
plt.plot(x, A, label='Low', color='blue')
plt.plot(x, B, label='Medium', color='green')
plt.plot(x, C, label='High', color='red')
plt.title('Fuzzy Membership Functions')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()
plt.show()
```

# Union of A and B

```python
A_or_B = np.maximum(A, B)

plt.figure(figsize=(10, 6))
plt.plot(x, A, label='A (Low)', color='blue')
plt.plot(x, B, label='B (Medium)', color='green')
plt.plot(x, A_or_B, label='A OR B', color='purple', linestyle='--')
plt.title('Union (A OR B)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()
plt.show()
```

# Intersection of A and B

```python
A_and_B = np.minimum(A, B)

plt.figure(figsize=(10, 6))
plt.plot(x, A, label='A (Low)', color='blue')
plt.plot(x, B, label='B (Medium)', color='green')
plt.plot(x, A_and_B, label='A AND B', color='orange', linestyle='--')
plt.title('Intersection (A AND B)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()
plt.show()
```

# Complement of A

```python
A_not = 1 - A

plt.figure(figsize=(10, 6))
plt.plot(x, A, label='A (Low)', color='blue')
plt.plot(x, A_not, label='NOT A', color='gray', linestyle='--')
plt.title('Complement (NOT A)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()
plt.show()
```

# Difference of A - B

```python
A_minus_B = np.maximum(0, A - B)

plt.figure(figsize=(10, 6))
plt.plot(x, A, label='A (Low)', color='blue')
plt.plot(x, B, label='B (Medium)', color='green')
plt.plot(x, A_minus_B, label='A - B', color='purple', linestyle='--')
plt.title('Difference (A - B)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()
plt.show()
```

# Commutative Property

```python
A_or_B_commutative = np.maximum(B, A)
A_and_B_commutative = np.minimum(B, A)
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x, A_or_B, label='A OR B', color='purple')
plt.plot(x, A_or_B_commutative, label='B OR A', linestyle='--', color='purple')
plt.title('Commutative Property (Union)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, A_and_B, label='A AND B', color='orange')
plt.plot(x, A_and_B_commutative, label='B AND A', linestyle='--', color='orange')
plt.title('Commutative Property (Intersection)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

# Associative Property

```python
A_or_B_or_C = np.maximum(A_or_B, C)
A_or_B_or_C_assoc = np.maximum(A, np.maximum(B, C))
A_and_B_and_C = np.minimum(A_and_B, C)
A_and_B_and_C_assoc = np.minimum(A, np.minimum(B, C))
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x, A_or_B_or_C, label='(A OR B) OR C', color='purple')
plt.plot(x, A_or_B_or_C_assoc, label='A OR (B OR C)', linestyle='--', color='purple')
plt.title('Associative Property (Union)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, A_and_B_and_C, label='(A AND B) AND C', color='orange')
plt.plot(x, A_and_B_and_C_assoc, label='A AND (B AND C)', linestyle='--', color='orange')
plt.title('Associative Property (Intersection)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
A = np.sin(x)
B = np.cos(x)
C = np.tan(x)

A_and_B_or_C = np.minimum(A, np.maximum(B, C))
A_and_B_or_C_combined = np.maximum(np.minimum(A, B), np.minimum(A, C))
A_or_B_and_C = np.maximum(A, np.minimum(B, C))
A_or_B_and_C_combined = np.minimum(np.maximum(A, B), np.maximum(A, C))
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x, A_and_B_or_C, label='A AND (B OR C)', color='purple')
plt.plot(x, A_and_B_or_C_combined, label='(A AND B) OR (A AND C)', linestyle='--', color='purple')
plt.title('Distributive Property (AND over OR)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, A_or_B_and_C, label='A OR (B AND C)', color='orange')
plt.plot(x, A_or_B_and_C_combined, label='(A OR B) AND (A OR C)', linestyle='--', color='orange')
plt.title('Distributive Property (OR over AND)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

# Idempotent Property

```python
A_idempotent_or = np.maximum(A, A)
A_idempotent_and = np.minimum(A, A)
plt.figure(figsize=(10, 6))

plt.plot(x, A, label='A', color='blue')
plt.plot(x, A_idempotent_or, label='A OR A', linestyle='--', color='blue')
plt.title('Idempotent Property (OR)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()

plt.figure(figsize=(10, 6))
plt.plot(x, A, label='A', color='blue')
plt.plot(x, A_idempotent_and, label='A AND A', linestyle='--', color='blue')
plt.title('Idempotent Property (AND)')
```
